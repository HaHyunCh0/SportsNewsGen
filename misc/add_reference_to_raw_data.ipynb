```python
# away win
awayteam_win_win = data[(data['FTHG'] < data['FTAG']) & (data['HTHG'] < data['HTAG'])] #1383
awayteam_draw_win = data[(data['FTHG'] < data['FTAG']) & (data['HTHG'] == data['HTAG'])] #884
awayteam_lose_win = data[(data['FTHG'] < data['FTAG']) & (data['HTHG'] > data['HTAG'])] #150
```

```python
# draw
hometeam_win_draw = data[(data['FTHG'] == data['FTAG']) & (data['HTHG'] > data['HTAG'])] #410
hometeam_lose_draw = data[(data['FTHG'] == data['FTAG']) & (data['HTHG'] < data['HTAG'])] #387
awayteam_win_draw = data[(data['FTHG'] == data['FTAG']) & (data['HTHG'] < data['HTAG'])] # 387
awayteam_lose_draw = data[(data['FTHG'] == data['FTAG']) & (data['HTHG'] > data['HTAG'])] #410
all_draw = data[(data['FTHG'] == data['FTAG']) & (data['HTHG'] == data['HTAG'])] # 1272
```

```python
# home lose
hometeam_win_lose = data[(data['FTHG'] < data['FTAG']) & (data['HTHG'] > data['HTAG'])] # 150
hometeam_draw_lose = data[(data['FTHG'] < data['FTAG']) & (data['HTHG'] == data['HTAG'])] #884
hometeam_lose_lose = data[(data['FTHG'] < data['FTAG']) & (data['HTHG'] < data['HTAG'])] #1383

# away lose
awayteam_win_lose = data[(data['FTHG'] > data['FTAG']) & (data['HTHG'] < data['HTAG'])] #197
awayteam_draw_lose = data[(data['FTHG'] > data['FTAG']) & (data['HTHG'] == data['HTAG'])] #1264
awayteam_lose_lose = data[(data['FTHG'] > data['FTAG']) & (data['HTHG'] > data['HTAG'])] #2342
```

```python
dataframes = {
    0:hometeam_win_win,1:awayteam_lose_lose,2:hometeam_draw_win,3:awayteam_draw_lose,4:hometeam_lose_win,
    5:awayteam_win_lose,6:awayteam_win_win,7:hometeam_lose_lose,8:awayteam_draw_win,9:hometeam_draw_lose,
    10:awayteam_lose_win,11:hometeam_win_lose,12:hometeam_win_draw,13:awayteam_lose_draw,14:hometeam_lose_draw,
    15:awayteam_win_draw,16:all_draw
}
```

```python
path_dir = 'templates'
file_list = os.listdir(path_dir)
file_list1 = natsort.natsorted(file_list)
```

```python
pd.set_option('mode.chained_assignment',  None)
```

```python
new_df = pd.DataFrame()
files_cnt = len(file_list1)

for i in range(files_cnt):
    filename = 'templates/' + file_list1[i]
    f = open(filename)
    num_types = len(f.readlines())
    df = dataframes[i].reset_index()
    
    df["FTHG"] = df["FTHG"].astype(str)
    df["FTAG"] = df["FTAG"].astype(str)
    df["HTHG"] = df["HTHG"].astype(str)
    df["HTAG"] = df["HTAG"].astype(str)
    df["realdate"] = df["realdate"].astype(str)
    
    if filename.split('_')[2] == 'home':
        df["bias"] = 'home'
    elif filename.split('_')[2] == 'away':
        df["bias"] = 'away'
    else:
        df["bias"] = 'neutral'

    if i != files_cnt - 1:
        if i % 2 == 0:
            df_len = int(len(df) / 2) + 1 if len(df) % 2 != 0 else int(len(df) / 2)
        else:
            df_len = int(len(df) / 2)
    else:
        df_len = len(df)
    print(df_len)
    for j in range(df_len):
        format=""
        line = random.randint(1,num_types)
        format = linecache.getline(filename,line)
        try:
            format1 = format.replace("[HomeTeam]",df.loc[j,"HomeTeam"])
            format2 = format1.replace("[AwayTeam]",df.loc[j,"AwayTeam"])
            format3 = format2.replace("[FTHG]",df.loc[j,"FTHG"])
            format4 = format3.replace("[FTAG]",df.loc[j,"FTAG"])
            format5 = format4.replace("[HTHG]",df.loc[j,"HTHG"])
            format6 = format5.replace("[HTAG]",df.loc[j,"HTAG"])
            format7 = format6.replace("[DateTime]",str(df.loc[j,"realdate"]))
        except Exception as e:
            print(i, line)
            pass
       
        df.loc[j,"article"] = format7
    new_df = pd.concat([new_df, df])
```

```python
new_df1 = new_df.dropna(subset=['Season','article'],axis=0)
```
